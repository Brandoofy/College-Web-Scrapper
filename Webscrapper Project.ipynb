```python
#  Imports all the librarys we are using for the webscraping
import requests
from bs4 import BeautifulSoup
```

```python
#request gets the website 
#mozillo is used to bypass python blocker
x = requests.get('https://www.niche.com/colleges/search/best-colleges/', headers={"user-agent":"Mozilla/5.0"})

#we created the soup var and used the html.parser method to store the html data
soup = BeautifulSoup(x.content, "html.parser") 

#div is html tag that is under college/class is list of elements
results = soup.find('div', {'class':'platform__wrapper'})  
store= results.findAll('li', class_='search-results__list__item') #li is html list of colleges 
# this class type of html shows what type is for college reviews/ class="search-result-feature__body">
#store all the html data into a str then print 
y = (str(store))
print(y)
```

```python
university_names = soup.findAll('h2', class_='MuiTypography-root MuiTypography-titleMedium MuiLink-root MuiLink-underlineHover nss-w5w7xf' ) #This is the reference to the html where the college names are stored

for i in university_names: #This For Loops Gets the raw text of the University names
    print(i.text)
```

```python
lst_of_uni = [name.text.upper() for name in university_names]
print(lst_of_uni)
```

```python
#grabs text from the refrence 
info = soup.findAll('p', class_= 'MuiTypography-root MuiTypography-body2 nss-1l1273x')
attributes = []

for l in info:
    attributes.append(l.text)

print(attributes)
```
